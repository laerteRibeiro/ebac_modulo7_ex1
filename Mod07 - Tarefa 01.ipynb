````python
cont_categorias = []
for variavel in df.columns:
    qtd = df[variavel].value_counts().count()
    cont_categorias.append(qtd)
    
cont_categorias
````

````python
dados['qtd_categorias'] = cont_categorias
dados
````

````python
nao_dummy = ['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']
````

````python
df_dummy = pd.get_dummies(df, columns=nao_dummy, drop_first=True)
df_dummy.head(2)
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df_dummy['sexo'] = df_dummy['sexo'].map({'M': 0, 'F': 1})
df_dummy['posse_de_veiculo'] = df_dummy['posse_de_veiculo'].map({'N': 0, 'Y': 1})
df_dummy['posse_de_imovel'] = df_dummy['posse_de_imovel'].map({'N': 0, 'Y': 1})
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
pd.crosstab(df_dummy['possui_email'], df_dummy['posse_de_veiculo'])
````

Acredito que a variável "posse_de_veiculo" seja mais poderosa, por ter mais equilíbrio entre as duas frequências de valores.
